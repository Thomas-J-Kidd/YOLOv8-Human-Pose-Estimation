````python
# Keypoint Detection Example
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
import os
from pathlib import Path
import contextlib
````

````python
# Define the path to the models folder
models_path = Path('../models').resolve()
model_file = models_path / 'yolo11n-pose.pt'
print("Models Path:", models_path)
print("Model File Path:", model_file)

# Check if the model file exists
if not model_file.exists():
    raise FileNotFoundError(f"Model file not found: {model_file}")

# Load YOLOv11n-pose model
model = YOLO(str(model_file))
````

````python
# Define the path to the image file
image_path = Path('human-image.jpeg').resolve()
print("Image File Path:", image_path)

# Check if the image file exists
if not image_path.exists():
    raise FileNotFoundError(f"Image file not found: {image_path}")
````

````python
# Load the image
img = cv2.imread(str(image_path))

# Suppress output
with contextlib.redirect_stdout(None):
    # Perform inference
    results = model.predict(img)

# Plot results with keypoints
plt.imshow(results[0].plot())
plt.axis('off')  # Hide axes
plt.show()
````
